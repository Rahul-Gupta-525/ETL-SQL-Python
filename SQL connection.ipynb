```python
# 1. Retrieve all customer details.
Q1 = " SELECT * FROM Customers; "

# 2. List all orders with their total amount and status.

Q2 = "SELECT order_id, total_amount FROM Orders;"

# 3. Find all products that belong to the ‘Electronics’ category.

Q3 = "SELECT product_name FROM ProductsWHERE category = 'Electronics';"

# 4. Find the total number of orders placed by each customer (customer_id and count).

Q4 = '''SELECT c.first_name, c.last_name, COUNT(*) Total_orders
FROM customers c
LEFT JOIN orders o ON c.customer_id = o.customer_id
GROUP BY 1,2;'''

# 5. Retrieve a list of all orders, including customer names and order status.

Q5 = '''SELECT o.order_id, c.first_name, o.status
FROM orders o
LEFT JOIN customers c ON c.customer_id = o.customer_id;'''

# 6. Find the total revenue (sum of all orders) generated. 

Q6 = '''SELECT ROUND(SUM(Total_amount),2) AS total_revenue
FROM orders;'''

# 7. Identify customers who have placed more than 3 orders.

Q7 = '''SELECT c.first_name
FROM customers c
JOIN orders o ON o.customer_id = c.customer_id
GROUP BY c.first_name
HAVING COUNT(*) > 3;'''


# 8. Retrieve all completed payments and include the associated order details.

Q8 = '''SELECT *
FROM orders o
JOIN payments p
ON o.order_id =  p.order_id
WHERE payment_status = "completed";'''

# 9. Find the top 5 customers who have spent the most on orders.
Q9 = '''SELECT c.first_name, c.last_name, ROUND(SUM(o.total_amount),2) AS Total_spent
FROM orders o
JOIN customers c ON c.customer_id = o.customer_id
GROUP BY o.customer_id
ORDER BY Total_spent DESC
LIMIT 5;'''

# 10. List all orders where the payment was either ‘Pending’ or ‘Failed’.

Q10 = '''SELECT o.order_id 
FROM orders o
JOIN payments p ON p.order_id = o.order_id
WHERE payment_status = 'Pending' or payment_status = 'failed';'''


# 11. Find the most popular product category based on the number of orders.

Q11 = '''SELECT p.category, COUNT(*) AS total_orders
FROM products p
JOIN orders o ON p.product_id = o.product_id
GROUP BY p.category
ORDER BY total_orders DESC
LIMIT 1;'''

# 12. Identify orders that were placed but have no successful payment recorded.

Q12 = '''SELECT o.order_id FROM orders o LEFT JOIN Payments p
ON o.order_id = p.order_id WHERE o.order_id NOT IN (SELECT order_id FROM payments WHERE Payment_status = "completed");'''


# 13. Calculate the average order value (AOV) per customer.
Q13 = '''SELECT c.Customer_id, COALESCE(ROUND(AVG(o.total_amount),2),0) AS AOV
FROM orders o 
RIGHT JOIN customers c ON c.customer_id = o.customer_id
GROUP BY customer_id;'''


# 14. Identify which payment method is most commonly used.
Q14 = '''SELECT payment_method
FROM payments
GROUP BY payment_method
ORDER BY COUNT(*)
LIMIT 1;'''

# 15.Determine the percentage of orders that were canceled.

Q15 = '''SELECT ROUND(((COUNT(*)/(SELECT COUNT(*) FROM orders))*100),2) AS percentage
FROM orders WHERE status = "Cancelled";'''

# 16. Find out the month in which the most orders were placed.

Q16 = '''SELECT monthname(order_date) AS month_, COUNT(*) As Total_orders
FROM orders
GROUP BY monthname(order_date) 
ORDER BY Total_orders DESC
LIMIT 1;'''
```
